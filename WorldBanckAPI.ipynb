# WBGAPI is a python package for accessing the World Bank's open data API. Available at https://pypi.org/project/wbgapi.

# Querying and Searching
## The info functions are the primary tool for searching the API. You can limit the response by calling info() with identifiers (a single value or lists) or by adding a keyword search.

- wb.source.info() # especially useful for seeing what databases are available
- wb.series.info()
- wb.economy.info('CAN')
- wb.economy.info(['USA', 'MEX', 'ARG'])
- wb.series.info(q='co2')
- wb.economy.info(q='korea')

```python
import wbgapi as wb   #  use wb as a namespace in all mwork
import pandas as pd   # optional but highly recommended for wbgapi
import matplotlib.pyplot as plt

%matplotlib inline
```

## The info functions are the primary tool for searching the API. You can limit the response by calling info() with identifiers (a single value or lists) or by adding a keyword search.

```python
wb.series.info()

wb.search('bonds')
```

## The info functions are the primary tool for searching the API. You can limit the response by calling info() with identifiers (a single value or lists) or by adding a keyword search.

```python
df=wb.data.DataFrame(['NY.GDP.PCAP.CD'], ['BRA', 'ARG', 'COL'], range(2000, 2020), index='time')
df
```

## Basic panda charts (using matplotlib) are straight forward. Of course, you can use any plotting package you like: seaborn, plotly, etc. Here is a plot of per-capita income for 3 countries and 2 decades:

```python
df.plot(figsize=(10, 6))
ax = df.plot(figsize=(10, 6), title="GDP OF Argentina, Brazil and Colombia")

plt.show()
```

# Aggregations
## Aggregations are easy in pandas. You simply use a summary function, something like this:

   df.mean()

## For this demonstration let's start by getting a small sample of data to aggregate

```python
df = wb.data.DataFrame('EG.ELC.ACCS.ZS', ['BRA', 'COL', 'US'], time=range(1990,2020))
df
```

```python
df.mean()
```

```python
combined = wb.data.DataFrame(['EG.ELC.ACCS.ZS', 'SP.POP.TOTL'], ['BRA', 'COL', 'US'],time=range(1990,2020), index=['series', 'economy'])
combined
```

```python
eco= wb.data.DataFrame(['EG.ELC.ACCS.ZS'], ['BRA', 'COL'],time=range(1990,2020), index=['time'])
eco
```

```python
eco.plot(figsize=(10, 6))
ax = eco.plot(figsize=(10, 6), title="Economy over time Brazil and Colombia")

plt.show()
```
